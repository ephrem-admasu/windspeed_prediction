#### Dynamic Time Warping Based Imputation

```python
from dtwi_analyzer import find_gaps, edtwbi
```

```python
wind_data_1 = df_1.fillna(value=0)['wind_speed'].values
filled_wind_data_1 = wind_data_1.copy()

gaps = find_gaps(wind_data_1)
```

```python
gaps
```

```python
max_gaps_idx = np.argmax([ix[1] - ix[0] for ix in gaps])
max_gaps_idx
```

```python
max_gap = gaps[6][1] - gaps[6][0]
max_gap
```

```python
print("From ", df_1.date.loc[gaps[6][0]], " to ", df_1.date.loc[gaps[6][1]])
```

#### Impute Missing Values

```python
for gap in gaps:
    if gap[1] - gap[0] < 0.3*len(wind_data_1):
        filled_wind_data_1[gap[0]:gap[1]] = edtwbi(wind_data_1, gap[0], gap[1])
```

### Defining the fitness function

```python
from LSSVMRegression import LSSVMRegression as LSSVM
from processing import create_features
from processing import split_data 
from sklearn.metrics import mean_squared_error as mse
```

```python
global X
def fitness(p):
    gamma = np.power(10, p[0]).astype(float)
    sigma = np.power(10, p[1]).astype(float)
    m = int(p[2])

    df = create_features(X, m)
    tr_arr, val_arr = split_data(df, int(.8*df.shape[0]))
    ytr = tr_arr['t']
    xtr = tr_arr.drop('t', axis=1)
    yval = val_arr['t']
    xval = val_arr.drop('t', axis=1)
    
    reg = LSSVM(kernel='rbf', gamma=gamma, sigma=sigma)
    reg.fit(xtr, ytr)
    yp = reg.predict(xval)
    return mse(yval, yp)
```
