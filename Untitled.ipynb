```python
def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = []
    for e in data:
        if(u-2*s < e < u+2*s):
            filtered.append(e)
        else:
            filtered.append(u)
    return pd.DataFrame(filtered)
```

```python
%matplotlib inline
import statsmodels.api as sm
from pandas.plotting import autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
#sm.graphics.tsa.plot_acf(df.values.squeeze(), lags=40)
lags = 100
plt.figure(figsize=(16, 4))
plot_acf(df.wind_speed.values.squeeze(), lags=lags)
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.title("")
for spine in plt.gca().spines.values():
    spine.set_visible(False)
# plt.savefig('corr.png', dpi=300)
plt.show()
```

```python
series = df.wind_speed.copy()
features = []

dataFrame = pd.DataFrame()
for i in range(lags, 0, -1):
    dataFrame['t-'+str(i)] = series.shift(i).values
    
dataFrame['t'] = series.values
dataFrame = dataFrame[lags+1:]

dataFrame.head()
```

```python
def plot_pred(yte, ypred, figsize=(12, 3), d = 100):
    plt.figure(figsize=figsize)
    plt.ylabel('Wind speed (m/sec)')
    plt.xlabel('Sample points (20 min.)')
    plt.plot(list(range(len(yte[0:d]))), yte[0:d])
    plt.plot(list(range(len(ypred[0:d]))), ypred[0:d], marker='*', color='red')
    plt.legend(['Actual values', 'Predicted values'])
```
